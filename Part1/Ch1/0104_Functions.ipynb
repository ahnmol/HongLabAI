### 다항식


$f(x) = x^3 - 2x + 1$

```python
def my_func(x):
    return x**3 - 2*x + 1

# x_values = []
# y_values = []
# for i in range(-10, 11):
#     x = float(i)
#     x_values.append(x)
#     y_values.append(my_func(x))

x_values = [float(i) for i in range(-10, 11)] # List comprehension
y_values = [my_func(x) for x in x_values]

print(x_values)
print(y_values)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x_values = np.linspace(-2, 2, 21)
y_values = my_func(x_values)

#print(x_values) # x_values.tolist()
#print(y_values) # y_values.tolist()

plt.plot(x_values, y_values, label="y = x³ - 2x + 1")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of y = x³ - 2x + 1")
plt.legend()

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values (avoiding zero since log(0) is undefined)
x = np.linspace(0.1, 10, 100)

# Compute log values
y = np.log(x)

# Plot the function
plt.plot(x, y, label="Natural Log Function")
plt.xlabel("x")
plt.ylabel("ln(x)")
plt.title("Logarithm Function Plot")
plt.legend()
plt.grid()
plt.show()
```

### 시그모이드 함수

Logistic sigmoid

$\sigma(x) = \frac{1}{1 + e^{-x}}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the derivative of sigmoid function
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Generate x values
x = np.linspace(-10, 10, 100)

# Compute sigmoid values and their derivatives
y = sigmoid(x)
y_derivative = sigmoid_derivative(x)

# Plot both functions
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Sigmoid Function σ(x)", color='blue')
plt.plot(x, y_derivative, label="Sigmoid Derivative σ'(x)", color='red', linestyle='--')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Sigmoid Function and Its Derivative")
plt.legend()
plt.grid()
plt.show()
```

[참고] 
1. [시그모이드 함수](https://en.wikipedia.org/wiki/Sigmoid_function)에는 여러가지가 있습니다. 예를 들어서, 위의 로지스틱 시그모이드는 범위가 0~1이고 하이퍼볼릭 탄젠트 함수는 범위가 -1~1입니다.

### 함수 객체

```python
def func1(x):
    return x ** 2

def func2(x):
    return 10 * x

def func3(x):
    return x + 3

f1 = func1
f2 = func2
f3 = func3

print(f1(3))         # x²
print(f2(f1(3)))     # 10x²
print(f3(f2(f1(3)))) # 10x² + 3
```

```python
x = 3

func_list = [f1, f2, f3]

for f in func_list:
    x = f(x)

print(x)
```
